```python
import numpy as np # numpy for math computations
import pandas as pd # pandas for dataframe
```

```python
playbyplay = pd.read_csv('play_by_play_2021.csv', low_memory=False) # converting csv into dataframe
playbyplay.shape # getting rows and columns
```

```python
playbyplay = playbyplay[playbyplay['play_type'] != 'no_play'] # taking out no_play rows
playbyplay.loc[playbyplay['qb_scramble'] == 1, 'play_type'] = 'pass' # converting scrambles into pass plays
playbyplay = playbyplay[playbyplay['season_type'] == 'REG'] # only regular season
```
