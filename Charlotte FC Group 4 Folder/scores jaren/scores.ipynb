```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import sklearn
```

```python
url = 'https://raw.githubusercontent.com/jarenrobbins/DTSC-3400/main/scores.csv'
scores = pd.read_csv(url)
```

```python
scores.head()
```

```python
home_games = scores[scores['Venue'] == 'Home']
```

```python
home_games.head()
```

```python
away_games = scores[scores['Venue'] == 'Away']
```

```python
away_games.head()
```

```python
bars = plt.bar(home_games['Opponent'], home_games['Attendance'], color=['red', 'green', 'green', 'green', 'green', 'red', 'green'])
plt.title('Home Games', color='white')
plt.xlabel('Opponent', color='white')
plt.ylabel('Attendance', color='white')
plt.xticks(rotation=60, color='white')
plt.yticks(color='white')
locs, labels = plt.yticks()
```

```python
bars1 = plt.bar(away_games['Opponent'], away_games['Attendance'], color=['red', 'red', 'red', 'red', 'blue', 'red', 'red',])
plt.title('Away Games', color='white')
plt.xlabel('Opponent', color='white')
plt.ylabel('Attendance', color='white')
plt.xticks(rotation=60, color='white')
plt.yticks(locs, labels, color='white')
```
